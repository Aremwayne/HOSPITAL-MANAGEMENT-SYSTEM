```python
import pymysql
import os
```

```python
PASSWORD = os.environ.get("MYSQL_PASSWORD")
```

```python
## Establish a connection between python and MYSQL
db = pymysql.connect(
    host ="localhost",
    port = 3306,
    user = "root",
    password = PASSWORD,
    database = "hospital_sys"
)
cursor = db.cursor()
if cursor:
    print("connection successful..")
    
```

```python
# Create a database
cursor.execute("create database Hospital_sys")
cursor.execute("show databases")
for i in cursor:
    print(i)
```

```python
# Create table in database
cursor.execute("use Hospital_sys")

cursor.execute ("""
 CREATE TABLE IF NOT EXISTS patients_records(
    patient_id  INT AUTO_INCREMENT PRIMARY KEY,
    first_name VARCHAR(50) NOT NULL,
    last_name VARCHAR(50) NOT NULL,
    date_of_birth DATE NOT NULL,
    gender ENUM("Male", "Female") NOT NULL,
    phone_no VARCHAR(20) NOT NULL,
    address TEXT NOT NULL
    )
""")
    
if cursor:
    print("Patients table created successfully!")
```

```python
cursor.execute("""
  CREATE TABLE IF NOT EXISTS doctors_records(
       doctors_id INT AUTO_INCREMENT PRIMARY KEY,
       first_name VARCHAR(50) NOT NULL,
       last_name VARCHAR(50) NOT NULL,
       specialization VARCHAR(100) NOT NULL,
       phone_no VARCHAR(20) NOT NULL,
       email VARCHAR(50) NOT NULL
     )
 """)
print("Doctors table created successfully!")
```

```python
cursor.execute("use Hospital_sys")

cursor.execute("""
  CREATE TABLE IF NOT EXISTS appointments(
       appointment_id INT AUTO_INCREMENT PRIMARY KEY,
       patient_id INT NOT NULL,
       doctors_id INT NOT NULL,
       appointment_date DATE NOT NULL,
       status ENUM ("Schedule", "Completed", "Cancelled"),
       FOREIGN KEY (patient_id) REFERENCES patients_records(patient_id),
       FOREIGN KEY (doctors_id) REFERENCES doctors_records(doctors_id)
    )
""")
print("Appointments table created successfully!")
```

```python
cursor.execute("use Hospital_sys")

cursor.execute("""
  CREATE TABLE IF NOT EXISTS billings(
       bill_id INT AUTO_INCREMENT PRIMARY KEY,
       patient_id INT NOT NULL,
       amount DECIMAL(10,2),
       bill_date DATE NOT NULL,
       payment_status ENUM ("Paid", "Unpaid"),
       FOREIGN KEY (patient_id) REFERENCES patients_records(patient_id)
    )
""")
print("Billings table created successfully!")
```

```python

def add_patient():
    first_name =input( "Enter your first name:")   
    last_name = input( "Enter your last name:")
    date_of_birth = input("Enter your Date of Birth (yyyy/mm/dd):")
    gender = input("Enter your Gender:")
    phone_no = input(" Enter your Phone no:")
    address = input("Enter your Address:")
    Age = int(input("Enter your Age:"))
    
    query = """ INSERT INTO patients_records
                (first_name, last_name, date_of_birth, gender, phone_no, address, Age)
                VALUES (%s, %s, %s, %s, %s, %s, %s)"""
    
    values = (first_name, last_name, date_of_birth, gender, phone_no, address, Age)
    
    cursor.execute(query, values)
    db.commit()
    print("patient added successfully!")  


def add_doctor():
    first_name = input( "Enter your first name:")
    last_name = input( "Enter your last name:")
    specialization = input( "Enter your Specialization:")
    phone_no =  input( "Enter your Phone no:")
    email = input( "Enter your Email:")
                      
    query = """ INSERT INTO doctors_records
                (first_name, last_name, specialization, phone_no, email)
                VALUES (%s, %s, %s, %s, %s)"""
    
    values = (first_name, last_name, specialization, phone_no,  email)
    
    cursor.execute(query, values)
    db.commit()
    print("Doctor added successfully!") 
          
          
def book_appointment():
    patient_id = int(input("Enter Patient ID:"))
    doctors_id = int (input("Enter Doctor ID:"))
    appointment_date = input("Enter appointment date(yy/mm/dd):")
    status = input("Enter Status (scheduled/Completed/Cancelled):")
    
    query = """ INSERT INTO appointments
                (patient_id, doctors_id, appointment_date, status)
                VALUES (%s, %s, %s, %s)"""
    
    values = (patient_id, doctors_id, appointment_date, status)
    
    cursor.execute(query, values)
    db.commit()
    print("Appointment booked successfully!")
    
    
def generate_bill():
    patient_id = int(input("Enter Patient ID:"))
    amount = float(input("Enter Amount: "))
    bill_date = input("Enter Bill Date (yyyy/mm/dd): ")
    payment_status = input("Enter Payment Status (Paid/Unpaid): ")
    
    
    query = """ INSERT INTO billings
                (patient_id, amount, bill_date, payment_status)
                VALUES (%s, %s, %s, %s)"""
    
    values = (patient_id, amount, bill_date, payment_status)
    
    cursor.execute(query, values)
    db.commit()
    print("Bill generated successfully!")
    
def view_patients():
    query = "SELECT * FROM patients_records"
    cursor.execute(query)
    results = cursor.fetchall()
          
    for row in results:
          print(row)
            
            
def view_doctors():
    query = "SELECT * FROM doctors_records"
    cursor.execute(query)
    results = cursor.fetchall()
          
    for row in results:
          print(row)          
        
        
def view_appointments():
    query = "SELECT * FROM appointments"
    cursor.execute(query)
    results = cursor.fetchall()
          
    for row in results:
          print(row) 
            
            
def view_bills():
    query = "SELECT * FROM billings"
    cursor.execute(query)
    results = cursor.fetchall()
          
    for row in results:
          print(row)
            
def search_patients():
    first_name = input("Enter patient first_name: ")
    query = "SELECT * FROM patients_records WHERE first_name = %s"
    cursor.execute(query, (first_name,))
    result = cursor.fetchall()
    if len(result) == 0:
        print("Records not Found")
    else:
        for i in result:
            print(i)
            
def delete_appointment():
    appointments_id = int(input("Enter appointment ID: "))
    query = "DELETE FROM appointments WHERE appointments_id = %s"
    cursor.execute(query, (appointments_id,))
    result = cursor.fetchall()
    db.commit()
    print("Appointment deleted successfully!")   
```
